```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ff['Gender'].value_counts()
```

```python
ff.Gender = ff.Gender.fillna('Male')
```

```python
ff.head(30)
```

```python
ff['Married'].value_counts()
```

```python
ff.Married = ff.Married.fillna('Yes')
```

```python
ff['Dependents'].value_counts()
```

```python
ff.Dependents = ff.Dependents.fillna('0')
```

```python
ff['Self_Employed'].value_counts()
```

```python
ff.Self_Employed = ff.Self_Employed.fillna('No')
```

```python
ff['LoanAmount'].value_counts()
```

```python
m=ff.LoanAmount.mean()
```

```python
m
```

```python
ff.LoanAmount = ff.LoanAmount.fillna(m)
```

```python
ff['Loan_Amount_Term'].value_counts()
```

```python
ff.Loan_Amount_Term = ff.Loan_Amount_Term.fillna(360)
```

```python
ff['Credit_History'].value_counts()
```

```python
ff.Credit_History = ff.Credit_History.fillna(1.0)
```

```python
total = ff.isnull().sum()
```

```python
total
```

```python
# Converting Categorical Variable into numeric:
```

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
le = LabelEncoder()
for i in var_mod:
    ff[i] = le.fit_transform(ff[i])
            
```

```python
ff.head()
```

```python
#Feature selection
```

```python
x = ff.iloc[:,1:12]
y = ff.iloc[:,12]
```

```python
x.shape
```

```python
y.shape
```

```python
x
```

```python
y.head()
```

```python
# Spliting Data
```

```python

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
x_test
```

```python
y_test
```

```python
x_train
```

```python
y_train
```

```python
# Decision Tree Model
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_predictions = model.predict(x_test)
```

```python
y_predictions
```

```python
y_test
```

```python
# Acuracy //Model Evaluation
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_predictions)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_predictions))
```

```python

# BUILDING LOGISTIC REGRESSION MODEL 
```

```python

from sklearn.linear_model import LogisticRegression
logistic_regression= LogisticRegression()
logistic_regression.fit(x_train,y_train)

```

```python
y_pred=logistic_regression.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
 
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
